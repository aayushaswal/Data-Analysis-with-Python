```python
#checking the shiping mode for order 
data["ship_mode"].value_counts().plot.pie()
```

```python
plt.figure(figsize=(7, 5))

# countplot: Show the counts of observations in each categorical bin using bars
sns.countplot(x='ship_mode', data=data, order=)

# Display the figure
plt.show()
```

From the above pie chart and barchart, we observe that 60% of people choose the standard class shipping mode. Only 5% of people use same-day delivery, and 15% opt for first-class delivery. The remaining individuals prefer the second-class delivery option.

```python
#order categorization
data["segment"].value_counts().plot.pie()
```

The entire set of orders is divided into three categories: 1) Consumer Goods, 2) Corporate, and 3) Home Office. Among these categories, 53% of the orders are from the Consumer Goods category, 27% are from the Corporate category, and the remaining orders are from the Home Office category.

```python
plt.figure(figsize=(5,25))

# countplot: Show the counts of observations in each categorical bin using bars
sns.barplot(x='sales', y='country',data=data)

# Display the figure
plt.show()
```

```python
plt.figure(figsize=(6,22))
sns.countplot(y="country",data=data)
plt.show()
```

From the above graphs, we can observe that the United States has a significantly large number of orders, approximately 10,000. However, the average amount spent per order is relatively low, around 300. The price variability in the US orders ranges from 240 to 350.

In contrast, Montenegro has the highest average sale value, with variations ranging from 18 to 2500. This suggests that people in the US are less inclined to make expensive purchases through superstores.

```python
#order by market
sns.countplot(x="market",data=data,)
```

The APAC market has the highest maximum number of orders, followed by LATAM, the US, and then the EU. Conversely, Canada has the fewest number of orders among these regions.

```python
#further categorization of market order
sns.countplot(x="market",data=data,hue='category')
```

Upon further categorization of market orders, it is evident that office supplies have the highest number of orders across all worldwide markets.

```python
# barplot: Sales Vs market
sns.barplot(x='market', y='sales',data=data,ci=None)

# Display the figure
plt.show()
```
