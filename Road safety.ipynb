```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data=pd.read_csv("UK_Accident.csv")
```

```python
data.head()
```

```python
print(data.info())
```

```python
data.describe()
```

```python
data.dropna(subset=['Location_Easting_OSGR'], inplace=True)
```

```python
print(data.info())
```

```python
sns.histplot(data['Accident_Severity'])
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.title('Distribution of Accident Severity')
plt.show()
```

```python
sns.countplot(data['Road_Type'])
plt.xlabel('Road_Type')
plt.ylabel('count')
plt.title('Distribution of road type')
plt.xticks(rotation=45)
plt.show()
```

```python
data_cas={'Casuality count' : data['Number_of_Casualties'],
          
    'Day_of_Week': data['Day_of_Week']
}
accidents_per_week = pd.DataFrame(data_cas)
accidents_per_week.groupby(by='Day_of_Week').sum().sort_values(by='Day_of_Week',ascending=True)
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#plt.figure(figsize=(12, 6))
data['Date'] = pd.to_datetime(data['Date'])
data['Year']= data['Date'].dt.year
accidents_by_year = data['Year'].value_counts().sort_index()
sns.lineplot(x=accidents_by_year.index, y=accidents_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Accidents Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Light_Conditions', y="Number_of_Casualties",data=data)
plt.xlabel('Light_Conditions')
plt.ylabel('Number_of_Casualties')
plt.title("Light_Conditions V/S Number_of_Casualties")
plt.xticks(rotation=46)
plt.show()
```
