```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load dataset

athletes = pd.read_csv("athlete_events.csv")
region = pd.read_csv("noc_regions.csv")
```

```python
athletes.head()
```

```python
region.head()
```

```python
#join the dataframe

data = athletes.merge(region,how = 'left',on ='NOC')
```

```python
data.head()
```

```python
data.shape
```

```python
#column name consistent

data.rename(columns={'region' : 'Region','notes': 'Notes'}, inplace=True)
```

```python
#Top Participating countries

top_10_countries = data.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(12,6))
plt.title("Overall participation of countries")
sns.barplot(x=top_10_countries.index, y= top_10_countries, palette='Set2' )
plt.show()
```

Graph showing the top 10 participating countries in the history of the Olympics, where the United States has the maximum number of participants (17,847)







```python
# age distribution of the participant

plt.figure(figsize=(12,6))
plt.title('Age distribution of participant')
plt.xlabel('Age')
plt.ylabel('Number of participant')
plt.hist(data.Age, bins =np.arange(10,80,2), color = 'orange', edgecolor = 'white' )
plt.show()
```

```python
Maximum no of participant age between 18 to 30 year
```

```python
#total medal

data.Medal.value_counts()
```

```python
#total number of female atheletes in each olympic

female_atheletes = data[(data.Sex == 'F') & (data.Season == 'Summer')][['Sex','Year']]
female_atheletes =female_atheletes.groupby('Year').count().reset_index()
female_atheletes.tail()
```

```python
female_participation = data[(data.Sex == 'F') & (data.Season == 'Summer')]
sns.set(style ="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=female_participation, palette='Spectral')
plt.title("Female athelete participation")
plt.show()
```

```python
part = female_participation.groupby('Year')["Sex"].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title("Plot of female athelete over time")
plt.show()
```

Graph indicate the participation of female atheletes over the year

```python
#gold medal atheletes

gold_medal = data[(data.Medal == 'Gold')]
gold_medal.head()
```

```python
#take only the value that are different from Nan
gold_medal= gold_medal[np.isfinite(gold_medal['Age'])]
```

```python
#goldmedal beyond 60

gold_medal['ID'][gold_medal['Age'] > 60].count()
```

```python
sport_event = gold_medal['Sport'][gold_medal['Age']>60]
sport_event
```

```python
#plot for sporting event
plt.figure(figsize=(10,6))
plt.tight_layout()
sns.countplot(sport_event)
plt.title('Gold medal for athelete age more than 60 years')
plt.show()
```

```python
#gold medal for each country

gold_medal.Region.value_counts().reset_index(name = 'Medal').head()
```

```python
total_gold_medal=gold_medal.Region.value_counts().reset_index(name = 'Medal').head()
g = sns.catplot(x="index", y="Medal", data = total_gold_medal,
               height= 5, kind ="bar",palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of medals")
plt.title("Gold medal per country")
plt.show()
```
