```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].median(),inplace=True)
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0],inplace=True)
```

```python
titanic['Fare'].fillna(titanic['Fare'].mean(),inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic_new=titanic.dropna(axis='index')
```

```python
titanic_new.isnull().sum()
```

```python
titanic_new.info()
```

```python
sns.countplot(x='Survived', data=titanic)
plt.show()
```

```python
survival_rate=titanic['Survived'].value_counts(normalize=True)*100
print(f"survival rate for titanic is {survival_rate}")
```

```python
sns.countplot(x="Sex",data=titanic,hue='Survived')
plt.show()
```

```python
sns.countplot(x="Sex",data=titanic,hue='Pclass')
plt.show()
```

```python
sns.countplot(x="Pclass",data=titanic,hue='Survived')
plt.show()
```

```python
fig=px.box(data_frame=titanic,y='Age',x='Survived')
fig.show()
```

```python
fig=px.box(data_frame=titanic,y='Fare',x='Survived')
fig.show()
```

```python
corr_mat=titanic.corr()
sns.heatmap(corr_mat,annot=True,cmap="rocket_r",)
```

```python
sns.countplot(x="Embarked",data=titanic,hue='Survived')
plt.show()
```

```python
sns.boxplot(data=titanic,x="Embarked",y="Age")
plt.show()
```

```python
sns.catplot(data=titanic,kind="violin",split=True,y="Age",x="Embarked")
```

```python
fig = px.violin(titanic, x='Pclass', y='Age', box=True, points='all')
fig.update_layout(
    xaxis_title='Passenger Class',
    yaxis_title='Age',
    title='Distribution of Age across Passenger Classes'
)

fig.show()
```

```python
# Create the scatter plot
fig = px.scatter(titanic, x='Age', y='Fare', color='Survived', title='Scatter Plot: Age vs Fare')

# Update layout
fig.update_layout(
    xaxis_title='Age',
    yaxis_title='Fare'
)

# Show the plot
fig.show()
```
