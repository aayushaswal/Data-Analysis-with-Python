# Exploratory Data Analysis


```python
# plotting a bar chart for Gender and it's count
ax= sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python

# plotting a bar chart for gender vs total amount

sales= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x='Gender', y="Amount", data= sales)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
#on basis of age

ax1 = sns.countplot(x='Age Group', hue='Gender', data=df)
for bar in ax1.containers:
    ax1.bar_label(bar)
    
```

```python
# Total Amount vs Age Group

sale_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x='Age Group',y='Amount',data=sale_age)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
# total number of orders from top 10 states

order_state= df.groupby(['State'],as_index=False)['Orders'].count().sort_values(by='Orders', ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(x='State', y='Orders', data=order_state)
plt.xticks(rotation=45)
plt.show()
```

```python
# total amount/sales from top  states

sale_state= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(x='State', y='Amount', data=sale_state)
sns.lineplot(x='State', y='Amount', data=sale_state)
plt.xticks(rotation=45)
plt.title("Statewise Sales")
plt.show()
```

```python
plt.figure(figsize=(5,5))
ax2 = sns.countplot(data=df, x= 'Marital_Status')
for i in ax2.containers:
    ax2.bar_label(i)
```

```python
#Sales by marital status and gender

sale_mg= df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(5,5))
sns.barplot(x='Marital_Status', y='Amount', data = sale_mg, hue='Gender' )
plt.show()
```

From above graphs we can see that most of the buyers are unmarried (women) and they have high purchasing power

```python
#occupation

plt.figure(figsize=(20,6))
oc=sns.countplot(data=df, x='Occupation')
for i in oc.containers:
    oc.bar_label(i)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(20,6))
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
plt.show()
```

```python
# Product Category
plt.figure(figsize=(20,7))
pc = sns.countplot(data = df, x = 'Product_Category')
plt.xticks(rotation = 45)
for i in pc.containers:
    pc.bar_label(i)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(20,7))
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
plt.xticks(rotation = 45)
plt.show()
```

Unmarried women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing, Electronics and Footwear
